```python
import pandas as pd 
import numpy as np
```

```python
%%time
df = pd.read_parquet('ic_normalised_final.parquet')
df
```

```python
%%time
df_gene_disease = pd.DataFrame(df,columns=['Orpha_ID','gene_symbol','weights'])
df_phen_gene = pd.DataFrame(df,columns=['hpo_id','gene_symbol'])
df_phen_gene
```

```python
%%time
df_gene_disease_dict = df_gene_disease.groupby(['gene_symbol'])['Orpha_ID'].apply(lambda grp: list(grp.value_counts().index)).to_dict()
df_phen_gene_dict = df_phen_gene.groupby(['hpo_id'])['gene_symbol'].apply(lambda grp: list(grp.value_counts().index)).to_dict()
```

```python
def cleanup(row):
    oid = []
    for x in row[1][1:-1].split(sep=','):
        try:
            oid.append(int(x.replace(' ','')))
        except:
            pass
    if oid == []:
        return -1
    return oid
test_df = pd.read_csv('ground_truth.csv')
test_df['phenotype']=test_df.apply(lambda row: [x.replace("'",'').replace(' ','') for x in row[0][1:-1].split(sep=',')],axis=1)
test_df['gene']=test_df.apply(lambda row: [x.replace("'",'').replace(' ','') for x in row[2][1:-1].split(sep=',')],axis=1)
test_df['oid'] = test_df.apply(cleanup,axis=1)
```

```python
df_phen_gene_dict['HP:0010301']
```

```python
%%time
def prediction(row):
    possible_genes ={}
    phenotypes = set(row[0])
    for phen in phenotypes:
#         if the phenotype from the test case doesnt exist then we add an empty list instead of removing the test case completely
        try:
            genes = df_phen_gene_dict[phen]
        except:
            genes = [] 
        for x in genes:
            if x not in possible_genes.keys():
                possible_genes[x] = 1
            else:
                possible_genes[x] += 1
#     print(possible_genes)
#     print('----------------------------')
#     possible_genes = dict(sorted(possible_genes.items(), key=lambda x:x[1], reverse=True))
    try:
        max_occur = max(possible_genes.values())
        possible_genes = dict({(k, v) for k, v in possible_genes.items() if v == max_occur})
#     print(possible_genes)
        return possible_genes.keys()
    except:
        return -1
test_df['result'] = test_df.apply(prediction,axis=1)
test_df
```

```python
test_df_2 = test_df[test_df['result'] != -1]
test_df_2
```

```python
%%time
disease_length_occur =[]
def predict_d(row):
    possible_oid ={}
#     print('---------')
#     print(row[3])
    for gene in row[3]:
        oid = df_gene_disease_dict[gene]
        for x in oid:
            if x not in possible_oid.keys():
                possible_oid[x] = 1
            else:
                possible_oid[x] += 1
    possible_oid = dict(sorted(possible_oid.items(), key=lambda x:x[1], reverse=True))
    try:
        max_occur = max(possible_oid.values())
#         print(possible_oid)
        possible_oid = dict({(k, v) for k, v in possible_oid.items() if v == max_occur})
#         print(possible_oid)
#         print(len(possible_oid))
        disease_length_occur.append(len(possible_oid))
#         print('___________________')
        return possible_oid.keys()
    except:
        return -1
test_df_2['disease_pred'] = test_df_2.apply(predict_d,axis=1)
```
